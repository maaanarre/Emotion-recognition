```python
frame_count = 0  # Initialisation du compteur de frames
while True:
    # Lecture de la vidéo ou du flux webcam
    
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    if frame_count % 2 != 0:  # Sauter une frame sur deux
        continue
    # effet mirroir 
    frame = cv2.flip(frame, 1)

    # Redimensionnement pour accélérer le traitement
    frame = cv2.resize(frame, (640, 480))

    # Chargement du Haar Cascade pour la détection de visages
    haar_cascade_path = "haarcascade_frontalface_default.xml"
    face_detector = cv2.CascadeClassifier(haar_cascade_path)

    # Conversion en niveaux de gris pour la détection
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Détection des visages
    num_faces = face_detector.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)

    # Traitement des visages détectés
    for (x, y, w, h) in num_faces:
        # Dessiner un rectangle autour du visage
        cv2.rectangle(frame, (x, y - 50), (x + w, y + h + 10), (0, 255, 0), 4)

        # Préparer le visage pour le modèle d'émotion
        roi_gray_frame = gray_frame[y:y + h, x:x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray_frame, (48, 48)), -1), 0)

        # Prédiction de l'émotion
        emotion_prediction = emotion_model.predict(cropped_img)
        maxindex = int(np.argmax(emotion_prediction))

        # Afficher l'émotion détectée sur le frame
        cv2.putText(frame, emotion_dict[maxindex], (x + 5, y - 20), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

    # Afficher la vidéo avec les annotations
    cv2.imshow('Emotion Detection', frame)

    # Sortie de la boucle si 'q' est pressé
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libération des ressources
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the model and weights (already done in your code)
# Assuming emotion_model and emotion_dict are defined

def preprocess_frame(frame):
    """Resize and preprocess the frame for emotion detection."""
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    resized_frame = cv2.resize(gray_frame, (48, 48))  # Resize to 48x48
    normalized_frame = resized_frame / 255.0  # Normalize to [0, 1]
    return np.expand_dims(np.expand_dims(normalized_frame, -1), 0)  # Add batch and channel dimensions

def predict_emotion(frame):
    """Predict emotion from a single frame."""
    preprocessed = preprocess_frame(frame)
    predictions = emotion_model.predict(preprocessed)
    emotion_index = np.argmax(predictions)
    emotion_label = emotion_dict[emotion_index]
    return emotion_label

# Test on a single frame
cap = cv2.VideoCapture(0)  # Change to video file path if needed

if not cap.isOpened():
    print("Error: Could not access the camera or video file.")
else:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("End of video or no frame captured.")
            break

        # Predict emotion
        emotion = predict_emotion(frame)

        # Display the result
        cv2.putText(frame, f"Emotion: {emotion}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.imshow("Emotion Detection", frame)

        # Exit when 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
